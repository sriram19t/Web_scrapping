```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url=requests.get('https://www.amazon.in/gp/bestsellers/books/').text
```

```python
url[0:100]
```

```python
#Beautiful soup parsing
parse=BeautifulSoup(url,'html.parser')
```

```python
parse.prettify()[0:100]
```

```python
# prase the content which is needed

scrape_content=parse.find('div',{
    'class':'a-cardui _cDEzb_card_1L-Yx'
})
```

```python
scrape_content.prettify()[0:10]
```

```python
#try to write a code in a where we could fetch all the Web scrapping data in a loop

Records_length=len(scrape_content('div',{
    'a-cardui _cDEzb_grid-cell_1uMOS expandableGrid p13n-grid-content'
    }))
scrape_data=scrape_content.find_all('div',{
    'a-cardui _cDEzb_grid-cell_1uMOS expandableGrid p13n-grid-content'
    })
```

```python

Book_name=np.array([])
Author_name=np.array([])
Book_image=np.array([])
Book_link=np.array([])
Book_rating_count=np.array([])
Book_rating=np.array([])
price=np.array([])
record=0

for i in range(0,Records_length,1):
  record=scrape_data[i]
  Book_name=np.append(Book_name,record.get_text('/n').split('/n')[1])
  Author_name=np.append(Author_name,record.get_text('/n').split('/n')[2])
  Book_image=np.append(Book_image,record.find('img')['src'])
  Book_link=np.append(Book_link,record.find('a',{
      'class':'a-link-normal'
  })['href'])
  if record.find('i') is not None:
    Book_rating=np.append(Book_rating,record.find('i').parent['title'][0:3])
    Book_rating_count=np.append(Book_rating_count,record.find('i').parent.find_all('span')[1].text)

  else:
    Book_rating=np.append(Book_rating,np.nan)
    Book_rating_count=np.append(Book_rating_count,np.nan)
  if record.find('span',{'class':'a-size-base'}) is not None:
    price=np.append(price,record.find('span',{'class':'a-size-base'}).find('span').text)
  else:
    price=np.append(price,record.find('span',{'class':'a-size-base a-color-price'}).child.text)







  # Book_rating=np.append(Book_rating,record.find('i').parent['title'])

```

```python
Amazon_Best_Selling=pd.DataFrame({
    'Book_name':Book_name,
    'Author_name':Author_name,
    'Book_image':Book_image,
    'Book_link':Book_link,
    'Book_rating':Book_rating,
    'Book_rating_count':Book_rating_count,
    'price':price

})
```
